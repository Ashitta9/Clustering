Scaling the data 

K Means Clustering 

```python
x=df
```

```python
x.head()
```

```python
from sklearn.cluster import KMeans
# Assuming k is your desired number of clusters
k = 3

# Create KMeans object and fit the scaled data
kmeans = KMeans(n_clusters=k, random_state=0)
x['Cluster'] = kmeans.fit_predict(x[['Alcohol', 'Color_Intensity']])

```

```python
centroids = kmeans.cluster_centers_
```

```python
sns.scatterplot(data=x, x='Alcohol', y='Color_Intensity', hue='Cluster', palette='Set1')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)
plt.show()
```

```python
# Calculate and print the inertia
inertia = kmeans.inertia_
print('Inertia:', inertia)
```

```python
k_range = range(1, 12)  # This generates a range from 1 to 11 (inclusive)
inertia_values = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x[['Alcohol', 'Color_Intensity']])
    inertia_values.append(kmeans.inertia_)

plt.plot(k_range, inertia_values, marker='o')
plt.title('Elbow Method for K')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()


```

As we can see from the above graph that there is turning like an elbow at k=3. So, we can say that the right number of cluster for the given datasets is 3.

Silhoutte score 

```python
from sklearn.metrics import silhouette_score
# Assuming x is your DataFrame
features_for_clustering = x[['Alcohol', 'Color_Intensity']]
cluster_labels = x['Cluster']

silhouette_avg = silhouette_score(features_for_clustering, cluster_labels)
print('Silhouette Score:', silhouette_avg)
```

Implementing K Means 

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_pred = kmeans.predict(x)

plt.scatter(x['Alcohol'], x['Color_Intensity'], c=y_pred) 

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker="*", s=250, c='red', edgecolors="k") 

plt.xlabel("Alcohol")
plt.ylabel("Color Intensity")
plt.title("KMeans k=3")
plt.show()

```

Agglomerative Clustering 

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Assuming x is your DataFrame
features_for_clustering = x[['Alcohol', 'Color_Intensity']]

```

```python
import scipy.cluster.hierarchy as sch

dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))

plt.title('Dendrogram')
plt.xlabel('categories')
plt.ylabel('Distance')
plt.show()
```

```python
# Perform Agglomerative clustering
agg_clustering = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
agg_labels = agg_clustering.fit_predict(features_for_clustering)
```

```python
agg_labels
```

Visulaization

Assuming agg_labels contains cluster labels 0,1,2,3

```python
# Convert features_for_clustering to a NumPy array for indexing
features_for_clustering_np = features_for_clustering.to_numpy()

# Plotting the results
for cluster_label in range(4):
    plt.scatter(
        features_for_clustering_np[agg_labels == cluster_label, 0],
        features_for_clustering_np[agg_labels == cluster_label, 1],
        s=100,
        label=f'Cluster {cluster_label + 1}'  # Corrected syntax and added label
    )

plt.title('Agglomerative Clustering')
plt.xlabel('Alcohol')
plt.ylabel('Color Intensity')
plt.legend()
plt.show()
```

Density-Based Spatial Clustering of Applications with Noise (DBSCAN)

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Assuming x is your DataFrame
features_for_clustering = x[['Alcohol', 'Color_Intensity']]
```

```python
# Assuming desired parameters for DBSCAN 
eps = 0.5  # The maximum distance between two samples 
min_samples = 5  # The number of samples (or total weight) 
```

```python
# Perform DBSCAN clustering
dbscan_clustering = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan_clustering.fit_predict(features_for_clustering)
```

```python
# Plotting the results
plt.scatter(
    features_for_clustering['Alcohol'],
    features_for_clustering['Color_Intensity'],
    c=dbscan_labels,
    cmap='viridis',
    s=100
)
plt.title('DBSCAN Clustering')
plt.xlabel('Alcohol')
plt.ylabel('Color Intensity')
plt.show()
```
